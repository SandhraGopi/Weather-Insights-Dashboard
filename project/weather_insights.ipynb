```python
column_mapping = {
    'Temperature_High (°C)': 'Temp_High_C',
    'Temperature_Low (°C)': 'Temp_Low_C', 
    'Humidity (%)': 'Humidity_Perc',
    'Wind_Speed (km/h)': 'Wind_Speed_kmh',
    'Precipitation (mm)': 'Precipitation_mm',
    'Visibility (km)': 'Visibility_km'
}
```

Rename columns using a dictionary.

```python
df = df.rename(columns=column_mapping)
```

Apply the rename dictionary.

```python
df.dtypes
```

Check data types.

```python
df['Date'] = pd.to_datetime(df['Date'])
```

Convert the Date column to datetime.

```python
df.dtypes
```

```python
df.duplicated().sum()
```

Checks duplicated data.


```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Month_Name'] = df['Date'].dt.month_name()
df['Day'] = df['Date'].dt.day
```

Extract time components from Date.

```python
df
```

```python
df['Temp_Range_C'] = df['Temp_High_C'] - df['Temp_Low_C']
```

Calculating the daily temperature range by subtracting the low temperature from the high temperature to analyze temperature variability.

```python
df
```

```python
df['Temp_Category'] = pd.cut(
    df['Temp_High_C'],
    bins=[-10, 10, 25, 40, 55],
    labels=['Cold', 'Mild', 'Warm', 'Hot']
)
```

Weather Categorization-Categorizing daily high temperatures into descriptive groups for easier weather pattern analysis and summary statistics.

```python
df
```

```python
df['AQI_Category'] = pd.cut(
    df['Air_Quality_Index'],
    bins=[0, 50, 100, 150, 200, 500],
    labels=['Good', 'Moderate', 'Poor', 'Unhealthy', 'Very Unhealthy']
)
```

Air Quality Categorization-Categorizing Air Quality Index values into health-based groups using standard EPA thresholds for clearer environmental impact analysis.

```python
df
```

```python
df.to_csv("cleaned_weather_data.csv", index=False)
```

```python
import matplotlib.pyplot as plt
plt.hist(df['Temp_High_C'],bins=20) #divides temperature into 20 equal sizes for a better plot 
plt.title("Distribution of High Temperature")
plt.xlabel("Temperature(Celsius)")
plt.ylabel("Frequency")
plt.show
```

Creating a histogram to visualize the frequency distribution of daily high temperatures across Cairo's weather dataset.

```python
df.groupby("Month_Name")["Temp_High_C"].mean().plot(kind="bar")
plt.title("Average High Temperature by Month")
plt.xlabel("Month")
plt.ylabel("Average Temperature(Celsius)")
plt.show()
```

Visualizing monthly temperature trends by showing the average high temperature for each month in a bar chart format.

```python
high_temp=df.groupby("Month_Name")["Temp_High_C"].mean()
low_temp=df.groupby("Month_Name")["Temp_Low_C"].mean()
temp=pd.DataFrame({
    "Average High Temperature":high_temp,
    "Average Low Temperature":low_temp
})
temp.plot(kind="bar")
plt.title("Avg High and Low Temperatures by Month")
plt.xlabel("Month")
plt.ylabel("Temperature(C)")
plt.legend()
plt.show()
```

Creating a grouped bar chart comparing monthly average high and low temperatures to visualize the temperature range throughout the year.
